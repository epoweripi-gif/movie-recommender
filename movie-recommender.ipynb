```python
# reading data and displaying table .head(n) function shows first n rows of a dataset five if ()

movies = pd.read_csv('/kaggle/input/datasets/organizations/grouplens/movielens-latest-small/movies.csv')
ratings = pd.read_csv('/kaggle/input/datasets/organizations/grouplens/movielens-latest-small/ratings.csv')

print(movies.head(10))
ratings.head(10)
movies.tail(16)
```

```python
# merge the ratings and movies tables
data = pd.merge(ratings, movies, on='movieId')
data.head(10)
```

```python
# make table of users vs movies they watched and what they rated them
user_movie = data.pivot_table(
    index='userId',
    columns='title',
    values='rating'
)

user_movie.head()
```

```python
# fill the non rated ones as 0
user_movie = user_movie.fillna(0)
user_movie.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(user_movie.T)
similarity.shape
```

```python
import numpy as np

def recommend(movie_name):
    movie_index = user_movie.columns.get_loc(movie_name)
    distances = similarity[movie_index]
    
    movie_list = sorted(
        list(enumerate(distances)),
        key=lambda x: x[1],
        reverse=True
    )[1:6]

    print("Recommended movies:")
    for i in movie_list:
        print(user_movie.columns[i[0]])
```

```python
recommend("Toy Story (1995)")
```

```python
import re

def remove_year(title):
    return re.sub(r"\s*\(\d{4}\)", "", title)

movies['clean_title'] = movies['title'].apply(remove_year)
```

```python

from difflib import get_close_matches

def find_movie(name):
    matches = get_close_matches(name, movies['clean_title'], n=1)
    return matches[0] if matches else  print("Movie not found!")
    return
    

```

```python
find_movie("toy story")
```
